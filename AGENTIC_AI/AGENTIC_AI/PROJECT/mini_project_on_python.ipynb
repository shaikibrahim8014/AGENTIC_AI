```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df=sns.load_dataset("titanic")
```

```python
print(df)
```

```python
#basic operations
print("Top data in  the  titanic  data\n")
print(df.head())
print("Least of the data in the titanic data sets")
print(df.tail())
print("\n\nshape of the titanic data sets")
print(df.shape)
```

```python
# statistics using age
print(df.describe())
print("\n\n max age \n\n",df['age'].max())
print("---------------------\n\n")
print("min age \n\n",df['age'].min())
print("---------------------\n\n")
print("mean age \n\n",df['age'].mean())
print("---------------------\n\n")
print("median age \n\n",df['age'].median())
print("---------------------\n\n")
print("standard deviation age \n\n",df['age'].std())
print("---------------------\n\n")
```

```python
plt.figure(figsize=(5,5))
plt.hist(df['age'],bins=20)
plt.xlabel('age')
plt.ylabel('count')
plt.title('age distribution')
plt.show()


```

```python
plt.figure(figsize=(5,5))
plt.plot(df['fare'])
plt.xlabel('fare')
plt.ylabel('count')
plt.title('fare distribution')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(df['age'],df['fare'])
plt.xlabel('age')
plt.ylabel('fare')
plt.title('age vs fare')
plt.show()
#
```

```python
plt.figure(figsize=(10,10))
plt.bar(df['age'],df['fare'])
plt.xlabel('age')
plt.ylabel('fare')
plt.title('age vs fare')
plt.show()
```
