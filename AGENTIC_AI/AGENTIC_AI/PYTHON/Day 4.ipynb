```python
#1d array=vector
a=np.array([1,2,3,4])
print(a)
```

```python
#2d array=matrice
b=np.array([[1,2,3],
           [4,5,6]])
print(b)
```

```python
c=np.array([
    [
        [1,2],
        [3,4]
    ],
        [
            [5,6],
            [5,4]
        ]
    ]
)
print(c)
```
